```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from scipy.stats import bootstrap

# Generare de date de exemplu
np.random.seed(42)  # Setăm semnul pentru reproducibilitate

# Date de exemplu pentru grup A și B
n = 30  # Număr de zile
group_A_conversions = np.random.binomial(100, 0.05, size=n)  # Rata de conversie de 5%
group_B_conversions = np.random.binomial(100, 0.07, size=n)  # Rata de conversie de 7%




# Funcție pentru a calcula intervalele de încredere
def calc_ci(data, confidence=0.95):
    return bootstrap((data,), np.mean, confidence_level=confidence).confidence_interval

# Grafic de BARE
plt.figure(figsize=(10, 6))
plt.bar(['Group A', 'Group B'], [np.mean(group_A_conversions), np.mean(group_B_conversions)],
        yerr=[calc_ci(group_A_conversions), calc_ci(group_B_conversions)],
        capsize=5, color=['blue', 'orange'], alpha=0.7, ecolor='gray', label='Confidence Interval (95%)')
plt.ylabel('Conversion Rate' , fontsize =  15)
plt.title('Comparison of Conversion Rates Between Group A and B' , fontsize =  20)
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()

# Grafic de LINIE
plt.figure(figsize=(10, 6))
plt.plot(range(n), group_A_conversions, label='Group A', marker='o')
plt.plot(range(n), group_B_conversions, label='Group B', marker='s')
plt.xlabel('Days', fontsize =  15)
plt.ylabel('Conversions' , fontsize =  15)
plt.title('Daily Conversion Rates for Group A and B' , fontsize =  20)
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
